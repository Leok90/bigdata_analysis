## 하이퍼파라미터 튜닝, RandomizedSearchCV

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
import numpy as np

rf = RandomForestClassifier(n_jobs=-1)

param_rf = {
    "n_estimators": np.arange(100, 1500, 100),
    "max_depth": np.arange(1, 31),
    "criterion": ["gini", "entropy"]
}

model = RandomizedSearchCV(
    estimator=rf,
    param_distributions=param_rf,
    n_iter=20,
    scoring="accuracy",
    verbose=10,
    n_jobs=1,
    cv=5
)

model.fit(X_train, y_train.iloc[:,1])
print(f"Best score: {model.best_score_}")

print("Best parameters set:")
best_parameters = model.best_estimator_.get_params()
for param_name in sorted(param_rf.keys()):
  print(f"\t{param_name}: {best_parameters[param_name]}")
```

## 투표 기반 분류기(Voting Classifier)

```python
from sklearn.ensemble import VotingClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_auc_score, accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
import numpy as np

log_clf = LogisticRegression(C=0.1, penalty='l1',	solver='saga')
rnd_clf = RandomForestClassifier(criterion='entropy', max_depth=21, n_estimators=500)
svm_clf = SVC(probability=True, C=1, kernel='rbf')

voting_clf = VotingClassifier(
    estimators=[('lr', log_clf), ('rf', rnd_clf), ('svc', svm_clf)],
    voting='soft')

idx_iter=0
proba_sum = np.zeros((X_test.shape[0],1))

kf = StratifiedKFold(n_splits=4, shuffle=True, random_state=71)
for tr_idx, va_idx in kf.split(X_train, y_train.iloc[:,1]):
  tr_x, va_x = X_train.iloc[tr_idx], X_train.iloc[va_idx]
  tr_y, va_y = y_train.iloc[tr_idx,1], y_train.iloc[va_idx,1]

  voting_clf.fit(tr_x, tr_y)
  pred = voting_clf.predict(va_x)

  # 반복 시 마다 정확도 측정 
  idx_iter += 1
  accuracy = np.round(accuracy_score(va_y,pred), 4)
  train_size = X_train.shape[0]
  test_size = X_test.shape[0]

  print('\n#{0} 교차 검증 정확도 :{1}, 학습 데이터 크기: {2}, 검증 데이터 크기: {3}'
          .format(idx_iter, accuracy, train_size, test_size))
  print('#{0} 검증 세트 인덱스:{1}'.format(idx_iter, va_idx))

  for clf in (log_clf, rnd_clf, svm_clf):
    clf.fit(tr_x, tr_y)
    y_pred = clf.predict(va_x)
    print(clf.__class__.__name__, accuracy_score(va_y, y_pred))

  # model prediction
  y_pred = pd.DataFrame(voting_clf.predict_proba(va_x)[:,1])
  # print(y_pred)

  # 평가지표인 ROC_AUC_SCORE 확인
  print('ROC-AUC SCORE : ' , roc_auc_score(va_y, y_pred))

  y_pred_final = pd.DataFrame(voting_clf.predict_proba(X_test)[:,1])
  proba_sum += y_pred_final
y_pred = proba_sum / 4
```
